```python
import yaml

# === Load existing YAML ===
with open("config.yaml", "r") as f:
    config = yaml.safe_load(f)

# === Edit all fields ===

config["lsm"] = {
    "func_name": "linear",
    "ndim": 300  
}

config["model"] = {
    "affine": True,
    "nflow": 20
}

config["optimizer"] = {
    "LR": 0.001  
}

config["output"] = {
    "save": True,
    "verbose": True
}

config["parameters"] = {
    "batch_size": 100,  # Change as needed
    "epochs": 50        # Number of training epochs
}

config["regularizer"] = {
    "LogDet": 1.0  # Regularization strength
}

# === Save updated YAML ===
with open("config.yaml", "w") as f:
    yaml.dump(config, f, default_flow_style=False)

#confirm
print("Updated config:")
print(yaml.dump(config, default_flow_style=False))
```
